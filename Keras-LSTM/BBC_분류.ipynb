```python
#제외어 (Stopword) 설정 

nltk.download('stopwords')
MY_STOP = set(nltk.corpus.stopwords.words('english'))

print('제외어:')
print(MY_STOP)
print('제외어 갯수:', len(MY_STOP))
```

```python
# 4번 블럭
# 데이터 보관 창고
original = []
articles = []
labels = []

print(type(original))
```

```python
with open('bbc-text.csv', 'r') as file:
    

    # 컬럼 이름 읽기
    reader = csv.reader(file)
    header = next(reader)
    print(header)

    # 기사 한줄씩 처리
    for row in reader:
        labels.append(row[0])
        original.append(row[1])

        # 제외어 삭제하기
        news = row[1]
        #print('전:', news)
        for word in MY_STOP:
            mask = ' ' + word + ' '
            news = news.replace(mask, ' ')
        #print('후:', news)

        articles.append(news)
print('처리한 기사 숫자:', len(articles))
```
